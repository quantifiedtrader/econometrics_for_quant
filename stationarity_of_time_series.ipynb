### Stationary and Non Stationay Time Series 

```python
#Import Libraries 
import yfinance as yf
import pandas as pd
from statsmodels.tsa.stattools import adfuller, kpss
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings("ignore")
```

```python
# Download data from Yahoo Finance for APLL stock for 1-minute timeframe
data = yf.download('AAPL', period='1d', interval='1m')
df = data.copy()

# Plot Adjusted Close Prices
plt.figure(figsize=(15, 5))
plt.plot(df.index, df['Adj Close'])
plt.title('Adjusted Close Prices')
plt.xlabel('Time')
plt.ylabel('Price')
plt.grid(True)
plt.show()

# Perform ADF, KPSS, and PP tests for Adjusted Close Prices
adf_result = adfuller(df['Adj Close'])
kpss_result = kpss(df['Adj Close'])
pp_result = adfuller(df['Adj Close'], regression='ct')

print("ADF Test Results for Adjusted Close Prices:")
print(f"Test Statistic: {adf_result[0]}")
print(f"P-value: {adf_result[1]}")
print("")

print("KPSS Test Results for Adjusted Close Prices:")
print(f"Test Statistic: {kpss_result[0]}")
print(f"P-value: {kpss_result[1]}")
print("")

print("Phillips-Perron Test Results for Adjusted Close Prices:")
print(f"Test Statistic: {pp_result[0]}")
print(f"P-value: {pp_result[1]}")
print("")
```

- The ADF test results indicate that the test statistic is -1.95 and the p-value is 0.31 for the Adjusted Close Prices. Since the p-value is greater than the significance level of 0.05, we fail to reject the null hypothesis of non-stationarity. This suggests that the Adjusted Close Prices may be non-stationary.

- The KPSS test results show that the test statistic is 1.63 and the p-value is 0.01 for the Adjusted Close Prices. Since the p-value is less than the significance level of 0.05, we reject the null hypothesis of stationarity. This suggests that the Adjusted Close Prices may be non-stationary.

- The Phillips-Perron test results indicate that the test statistic is -1.88 and the p-value is 0.67 for the Adjusted Close Prices. Since the p-value is greater than the significance level of 0.05, we fail to reject the null hypothesis of non-stationarity. This aligns with the results from the ADF test.

```python
# Calculate Difference for Adjusted Close Price
df['Diff_Adj_Close'] = df['Adj Close'].diff()

# Plot Difference of Adjusted Close Price
plt.figure(figsize=(15, 5))
plt.plot(df.index[1:], df['Adj Close'].diff().dropna())
plt.title('Difference of Adjusted Close Price')
plt.xlabel('Time')
plt.ylabel('Difference')
plt.grid(True)
plt.show()

# Perform ADF, KPSS, and PP tests for Difference of Adjusted Close Price
adf_result_diff = adfuller(df['Diff_Adj_Close'].dropna())
kpss_result_diff = kpss(df['Diff_Adj_Close'].dropna())
pp_result_diff = adfuller(df['Diff_Adj_Close'].dropna(), regression='ct')

print("ADF Test Results for Difference of Adjusted Close Price:")
print(f"Test Statistic: {adf_result_diff[0]}")
print(f"P-value: {adf_result_diff[1]}")
print("")

print("KPSS Test Results for Difference of Adjusted Close Price:")
print(f"Test Statistic: {kpss_result_diff[0]}")
print(f"P-value: {kpss_result_diff[1]}")
print("")

print("Phillips-Perron Test Results for Difference of Adjusted Close Price:")
print(f"Test Statistic: {pp_result_diff[0]}")
print(f"P-value: {pp_result_diff[1]}")
print("")
```

- The ADF test results indicate that the test statistic is -21.60 and the p-value is 0.0 for the Difference of Adjusted Close Price. Since the p-value is less than the significance level of 0.05, we reject the null hypothesis of non-stationarity. This suggests that the Difference of Adjusted Close Price is likely stationary.

- The KPSS test results show that the test statistic is 0.13 and the p-value is 0.1 for the Difference of Adjusted Close Price. Since the p-value is greater than the significance level of 0.05, we fail to reject the null hypothesis of stationarity. This aligns with the results from the ADF test, indicating that the Difference of Adjusted Close Price is stationary.

- The Phillips-Perron test results indicate that the test statistic is -15.39 and the p-value is 1.505e-22 for the Difference of Adjusted Close Price. Since the p-value is less than the significance level of 0.05, we reject the null hypothesis of non-stationarity. This further confirms that the Difference of Adjusted Close Price is stationary.-

```python
# Calculate Percentage Change
df['Percent_Change'] = df['Adj Close'].pct_change()

plt.figure(figsize=(15, 5))
plt.plot(df.index[1:], df['Adj Close'].pct_change().dropna())
plt.title('Percentage Change')
plt.xlabel('Time')
plt.ylabel('Change')
plt.grid(True)
plt.show()

# Perform ADF, KPSS, and PP tests for Percentage Change
adf_result_pct_change = adfuller(df['Percent_Change'].dropna())
kpss_result_pct_change = kpss(df['Percent_Change'].dropna())
pp_result_pct_change = adfuller(df['Percent_Change'].dropna(), regression='ct')

print("ADF Test Results for Percentage Change:")
print(f"Test Statistic: {adf_result_pct_change[0]}")
print(f"P-value: {adf_result_pct_change[1]}")
print("")

print("KPSS Test Results for Percentage Change:")
print(f"Test Statistic: {kpss_result_pct_change[0]}")
print(f"P-value: {kpss_result_pct_change[1]}")
print("")

print("Phillips-Perron Test Results for Percentage Change:")
print(f"Test Statistic: {pp_result_pct_change[0]}")
print(f"P-value: {pp_result_pct_change[1]}")
print("")
```

- The ADF test results indicate that the test statistic is -21.60 and the p-value is 0.0 for the Percentage Change. Since the p-value is less than the significance level of 0.05, we reject the null hypothesis of non-stationarity. This suggests that the Percentage Change is likely stationary.

- The KPSS test results show that the test statistic is 0.13 and the p-value is 0.1 for the Percentage Change. Since the p-value is greater than the significance level of 0.05, we fail to reject the null hypothesis of stationarity. This aligns with the results from the ADF test, indicating that the Percentage Change is stationary.

- The Phillips-Perron test results indicate that the test statistic is -15.39 and the p-value is 1.506e-22 for the Percentage Change. Since the p-value is less than the significance level of 0.05, we reject the null hypothesis of non-stationarity. This further confirms that the Percentage Change is stationary.
